### Histogram, Skew, Kurt
Biểu đồ Histogram là một dạng biểu đồ cột được sử dụng để mô tả trực quan sự phân bố tần suất cho tập dữ liệu.

Biểu đồ Histogram có thể cung cấp các thông tin:

* Trung tâm về mặt vị trí của tập dữ liệu;
* Độ phân tán của tập dữ liệu;
* Độ lệch của tập dữ liệu;
* Sự hiện diện của các giá trị ngoại lệ (outliers);
* Sự hiện diện của các yếu vị (mode) trong tập dữ liệu.

```python
data[['Dia', 'GDCD', 'Hoa', 'Li', 'Ngoai_ngu', 'Sinh', 'Su', 'Toan', 'Van']].hist()
plt.gcf().set_size_inches(20,8)
```

```python
data['Toan'].hist(bins=300)
plt.gcf().set_size_inches(20,8)
plt.xlabel("Toan")
plt.ylabel('Frequency')
plt.title('Toan Histogram')
```

```python
#tính độ lệch skewness, có thể lớn hơn 1, + là lệch phải-positive
data[['Dia', 'GDCD', 'Hoa', 'Li', 'Ngoai_ngu', 'Sinh', 'Su', 'Toan', 'Van']].skew()
```

* Sk>0, được gọi là độ lệch tích cực hay Positive Skewness): đuôi bên PHẢI dài hơn đuôi bên trái; giá trị LỚN (outliers) đẩy giá trị trung bình về phía CUỐI: Ngoại ngữ, Sinh, Sử

* Sk=0, được gọi là phân phối đối xứng hay Symetrical distribution: 

* Sk<0, được gọi là độ lệch tiêu cực hay Negative Skewness): đuôi bên TRÁI dài hơn đuôi bên phải; giá trị NHỎ (outliers) đẩy mean về phía ĐẦU: GDCD, Văn

![](http://thongke.cesti.gov.vn/images/DV_PTDL/TKMT-P3-HinhdangPP/Relationship_between_mean_and_median_under_different_skewness.jpg)

```python
#tính độ nhọn kurtosis, có thể lớn hơn 1, + là nhọn
data[['Dia', 'GDCD', 'Hoa', 'Li', 'Ngoai_ngu', 'Sinh', 'Su', 'Toan', 'Van']].kurt()
```

* Mesokurtic (Hệ số Kurt = 0)

* Platykurtic (Hệ số Kurt < 0): "Platy-" có nghĩa là "rộng" có Hóa, Lí, Toán

* Leptokurtic (Hệ số Kurt > 0): "Lepto-" có nghĩa là "mảnh mai", Ngoại ngữ, Sinh
![](http://thongke.cesti.gov.vn/images/DV_PTDL/TKMT-P3-HinhdangPP/Picture2.png)

### Correlation

```python
#vẽ heatmap, thể hiện hệ số tương quan giữa các môn học
cor = data[['Dia', 'GDCD', 'Hoa', 'Li', 'Ngoai_ngu', 'Sinh', 
      'Su', 'Toan', 'Van']].corr()
sns.heatmap(cor, annot=True, cmap =plt.cm.Reds )
plt.gcf().set_size_inches(15,8)
```

Corr(toán, lý) =0.67 khá cao, Corr(toán, hóa)= 0.62, Corr(địa, sử) = 0.6
==> Các môn cùng khối có corr cao hơn
