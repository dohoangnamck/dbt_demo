## Feature Scaling: Normalization (max-min scaling), Standardization (Z- scaling)

### Normalization (max-min scaling)

```python
# mô tả boxplot của 'Year working' và 'Age'
data[['Year working', 'Age']].plot(kind='box')
```

Kết quả rất khó hình dung để so sánh 

```python
df = data[['Year working', 'Age']]
```

Chuẩn hóa max-min đưa về dạng 0-1

```python
# Scaling the dataset (normalization):x_mms=(x-x_min)/(x_max - x_min)
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms.fit(df)
data_mms=mms.transform(df)
```

```python
# Convert to Dataframe
data_mms=pd.DataFrame(data_mms, columns=['Year working', 'Age'])
data_mms.head(10)
```

```python
data_mms.plot(kind='box')
plt.title('data after max_min normalization')
```

### Standardization (Z- scaling)


```python
""""
Chuẩn hóa theo Z-Score
Tính khoảng lệch cao nhất so với trung bình, 
Z-Score cho biết giá trị quan sát lệch so vs mean, bằng bnh lần Độ lệch chuẩn
"""
```

```python
#Standardization: x_stand = (x-mean(x))/StandardDeviation  
from sklearn.preprocessing import StandardScaler 
sc=StandardScaler()
data_stand=sc.fit_transform(df)
```

## Discretization: Rời rạc hóa dữ liệu

```python
data.head()
```

```python
"""
Đôi lúc những cái mô hình phân tích dữ liệu cần phải rời rạc hóa nó không phải
liên tục (vd phân lớp dữ liệu bằng cây quyết định, xếp loại vào các mức)
"""
```

```python
#Discretization using CUT method of pandas package 
# the bins [1, 2, 3, 4] indicate (1,2], (2,3], (3,4]
# Chia cột Year working thành 4 mức

cut_labels = ['Internship', 'Fresher', 'Junior', 'Senior'] 
cut_bins = [-1, 1, 2, 4, 7]  #note: 0: min, 7: max 
data['Year_Cat'] = pd.cut(data['Year working'], bins=cut_bins, labels=cut_labels)
```

```python
data
```

```python
#chia thành 5 cấp dựa vào Department 
data['Dep_Cat']=pd.cut(data['Department'],5, labels=['Sale', 'IT', 'HR', 'Admin', 'Analysis']) 
data.head(10)
```

```python
data['Year_Cat'].value_counts()
```
