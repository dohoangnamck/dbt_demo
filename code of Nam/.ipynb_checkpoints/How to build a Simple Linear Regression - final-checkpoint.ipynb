In simple linear regression we have only one dependent variable and one independent variable, so we need one training feature. this is the machine learning language training feature

so using x and y we need to find out m and c the training parameters so that if we have x we can predict or estimate y.

```python
#Import Libraries 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

```python
#Load dataset  
data=pd.read_csv('weightheight.csv')
```

```python
#Quick view about dataset
data
#height: inch; weight: pound
```

```python
# Edit columns
data.columns= ['height', 'weight']
```

```python
data.plot(kind='scatter', x='height', y='weight')
plt.show()
```

```python
data.plot(kind='box')
plt.show()
#cÃ³ thá»ƒ scale láº¡i náº¿u cáº§n thiáº¿t
```

```python
data.corr()  #Correlation Coefficients: há»‡ sá»‘ tÆ°Æ¡ng quan
```

```python
#change to dataframe variables: biáº¿n khung dá»¯ liá»‡u
height = pd.DataFrame(data['height']) 
weight = pd.DataFrame(data['weight'])
```

```python
X = height
y = weight
```

![](https://i.imgur.com/7hZ5NM6.jpg)
m which is the slope of the line

In Machine Learning: m is called coefficient; m and C are called training parameters 

```python
#Build Linear regression model 
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X, y)
```

```python
c = lr.intercept_
c
```

```python
m = lr.coef_
m
```

```python
lr.score(height, weight) 
```

```python
#Predict new value of Weight 
import numpy as np

height_new = 63                    # type the height value to predict weight 
height_new = np.array(height_new)
weight_predict = lr.predict(height_new.reshape(-1, 1))
print(weight_predict, "pounds")
```

In NumPy, -1 in reshape(-1) refers to an unknown dimension that the reshape() function calculates for you. It is like saying: â€œI will leave this dimension for the reshape() function to determineâ€. A common use case is to flatten a nested array of an unknown number of elements to a 1D array.

```python
#Predict more values 
X=([65,67,72]) 
X = pd.DataFrame(X) 
Y = lr.predict(X) 
Y = pd.DataFrame(Y) 
df = pd.concat([X,Y], axis=1, keys = ['height_new_inch', 'weight_predict_pound']) 
df
```

```python
#Visualize the result 
data.plot(kind='scatter', x='height', y='weight') 

#plotting the Regression Line  
plt.plot(height, lr.predict(height), color='red', linewidth=2) 

#plotting the predict values 
plt.scatter(height_new, weight_predict, color='black') 
plt.plot(X, Y, color = 'blue', linewidth=3)
plt.show()
```

This screen is `one dimensional` in psychic learning libraries the machine learning algorithms take `two dimensional` data so we need to reshape this x_train and x_test to two dimensional

```python
# reshape X_train and X_test to two dimensional
import numpy as np
X_train = np.array(X_train).reshape(-1,1)
X_train
```

```python
X_test = np.array(X_test).reshape(-1,1)
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

When we fit X_train and y_train to this linear regression model this is called the training that means we actually trained this model and after this fitting process is done that means the training is done. This lr model should have the training parameters m and c.

```python
c = lr.intercept_
c
# c lÃ  intercept: há»‡ sá»‘ cháº·n
```

```python
m = lr.coef_
m
# m is coefficient, slope of line
```

```python
print('R square: ', lr.score(X_train, y_train))
```

Now we already have m and c so if we have the x now we should be able to estimate the y.

```python
Y_pred_train = m*X_train + c
Y_pred_train.flatten()
```

But for scikit-learn library there is a predict formula so you can use directly the predict formula

we already fit X_train and y_train to the lr that means lr should be trained and it should be able to predict now

```python
y_pred_train1 = lr.predict(X_train)
y_pred_train1.flatten()
```

so you can see that the values are exactly the same, so you do not need to calculate like this you can simply use the predict function and it will calculate it for you.

```python
import matplotlib.pyplot as plt 
plt.scatter(X_train, y_train) 
plt.plot(X_train, y_pred_train1, color ='red') 
plt.xlabel("height") 
plt.ylabel("weight")
```

Now what to do with X_test and y_test, we used X_train and y_train to train
the model right here 

but the model have never seen X_test and y_test. so we can test this model using X_test now let's see the way it was able to predict for X_train if it can predict the same for X_test

```python
y_pred_test1 = lr.predict(X_test)
y_pred_test1 #.flatten()
```

```python
import matplotlib.pyplot as plt 
plt.scatter(X_test, y_test) 
plt.plot(X_test, y_pred_test1, color ='red') 
plt.xlabel("height") 
plt.ylabel("weight")
```

so look at it, it was able to find the best fit line for X_test as well though the linear regression model never have seen X_test so it can generalize or predict weight for any height. 

So this was a simple linear regression where we had only one training feature height to predict weight but it doesn't have to be on the one training feature. Linear regression can work when there are multiple training features as well and we call it multiple linear regression in my next video. 

```python
#other ways to calculate r2_score
from sklearn.metrics import r2_score
r2_score(y_train, y_pred_train1)
```
