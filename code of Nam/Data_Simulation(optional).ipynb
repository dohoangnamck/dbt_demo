```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Set random seed for reproducibility
random.seed(42)

# Generate random unique 12-digit App IDs
app_ids = random.sample(range(10**11, 10**12), 108000)

# Generate random unique 9-digit eKYC IDs
ekyc_ids = random.sample(range(10**8, 10**9), 108000)

# Define value pools for Source, Channel, and Ad platforms
source_pool = ['RB', 'Telesale', 'Facebook', 'Google', 'ZaloAdtima', 'AccessTrade', 'BAMBOO', 'GOTADI', 'VNDIRECT', 'CASHBAG', 'FPT_PLAY', 'MOMO', 'Universities']
channel_pool = ['RB', 'Telesale', 'Digital Direct Sale', 'Ecosystem', 'Partnership']
ad_platforms = ['Facebook', 'Google', 'ZaloAdtima', 'AccessTrade']

# Generate data for each attribute
crm_data = {
    'App_ID': random.sample(app_ids, 108000),
    'eKYC_ID': random.sample(ekyc_ids, 108000),
    'Source': [],
    'Channel': [],
    'Click_Ads': [],
    'Install_App': [],
    'eKYC': [],
    'account_created': [],
    'trans_date': []
}

# Generate values for Source, Channel, Click Ads, Install App, and eKYC
for i in range(108000):
    channel = random.choice(channel_pool)

    if channel == 'RB':
        source = 'RB'
    elif channel == 'Telesale':
        source = 'Telesale'
    elif channel == 'Digital Direct Sale':
        source = random.choice(ad_platforms)
    elif channel == 'Ecosystem':
        source = random.choice(['BAMBOO', 'GOTADI', 'VNDIRECT'])
    elif channel == 'Partnership':
        source = random.choice(['CASHBAG', 'FPT_PLAY', 'MOMO', 'Universities'])

    click_ads = random.uniform(datetime(2022, 1, 1), datetime(2023, 1, 1))
    install_app = None if random.random() <= 0.15 else random.uniform(click_ads, datetime(2023, 1, 1))
    ekyc = None if install_app is None or random.random() <= 0.27 else random.uniform(install_app, datetime(2023, 1, 1))
    account_created = None if ekyc is None or random.random() <= 0.32 else random.uniform(ekyc, datetime(2023, 1, 1))
    trans_date = None if account_created is None or random.random() <= 0.53 else random.uniform(account_created, datetime(2023, 1, 1))

    crm_data['Source'].append(source)
    crm_data['Channel'].append(channel)
    crm_data['Click_Ads'].append(click_ads)
    crm_data['Install_App'].append(install_app)
    crm_data['eKYC'].append(ekyc)
    crm_data['account_created'].append(account_created)
    crm_data['trans_date'].append(trans_date)
```

```python
# Generate random string with specified length
def generate_random_string(length):
    letters = '0123456789'#string.ascii_uppercase + string.ascii_lowercase
    return ''.join(random.choice(letters) for _ in range(length))

# Generate random Transaction_ID
def generate_transaction_id():
    return 'FT_' + generate_random_string(6)

# Generate random Transaction_Account
def generate_transaction_account():
    return '-'.join([generate_random_string(4) for _ in range(3)]) + '-' + generate_random_string(4)

# Generate random Customer ID
def generate_customer_id(ekyc_df, n):
    ekyc_customer_ids = ekyc_df[~(ekyc_df['CUSTOMER_ID']=='NULL')]['CUSTOMER_ID'].sample(frac=0.67, random_state=42).tolist() #ekyc_df['CUSTOMER_ID'].sample(frac=0.67, random_state=42).tolist()
    non_ekyc_customer_ids = [generate_random_string(6) for _ in range(n - len(ekyc_customer_ids))]
    return ekyc_customer_ids + non_ekyc_customer_ids

# Generate random Account_Category
def generate_account_category():
    categories = ['1001', '1002', '1003', '1004', '1005']
    return random.choice(categories)

# Generate random Channel based on Account_Category
def generate_channel(account_category):
    if account_category in ['1001', '1002']:
        return 'APP'
    else:
        channels = ['WEB', 'RB', 'PORTAL']
        return random.choice(channels)

# Generate random Transaction_Type
def generate_transaction_type():
    types = ['NapasTransfer', 'CitadTransfer', 'InternalPayment', 'BillPayment', 'TopupPayment', 'TutionFee', 'FacePay', 'TopupAccount', 'OpenDeposit']
    return random.choice(types)

# Generate random Transaction_Group based on Transaction_Type
def generate_transaction_group(transaction_type):
    if transaction_type in ['NapasTransfer', 'CitadTransfer', 'InternalPayment']:
        return 'TRANSFER'
    elif transaction_type in ['BillPayment', 'TopupPayment', 'TutionFee', 'FacePay']:
        return 'PAYMENT'
    elif transaction_type in ['TopupAccount', 'OpenDeposit']:
        return 'DEPOSIT'

# Generate random Transaction_Amount based on Transaction_Type
def generate_transaction_amount(transaction_type):
    if generate_transaction_group(transaction_type) == 'TRANSFER':
        return random.uniform(20000, 200000000)
    elif generate_transaction_group(transaction_type) == 'PAYMENT':
        return random.uniform(20000, 500000000)
    elif generate_transaction_group(transaction_type) == 'DEPOSIT':
        return random.uniform(1000000, 1000000000)

# Generate random Transaction_Amount based on Transaction_Type
def generate_transaction_dt():
  # Define the start and end datetimes
  start_datetime = datetime(2022, 1, 1, 0, 0, 0)
  end_datetime = datetime(2023, 1, 1, 0, 0, 0)

  # Generate a random number of seconds within the specified range
  random_seconds = random.randint(0, int((end_datetime - start_datetime).total_seconds()))

  # Add the random number of seconds to the start datetime
  random_datetime = start_datetime + timedelta(seconds=random_seconds)
  return random_datetime

# Create the DataFrame
def generate_dataframe(n, ekyc_df):
    data = {
        'Transaction_ID': [generate_transaction_id() for _ in range(n)],
        'Transaction_Account': [generate_transaction_account() for _ in range(n)],
        'Customer_ID': generate_customer_id(ekyc_df, n),
        'Account_Category': [generate_account_category() for _ in range(n)],
    }

    data['Channel'] = [generate_channel(category) for category in data['Account_Category']]
    data['Transaction_Type'] = [generate_transaction_type() for _ in range(n)]
    data['Transaction_Group'] = [generate_transaction_group(t_type) for t_type in data['Transaction_Type']]
    data['Transaction_Amount'] = [generate_transaction_amount(t_type) for t_type in data['Transaction_Type']]
    data['Transaction_DT'] = [generate_transaction_dt() for _ in range(n)]

    df = pd.DataFrame(data)
    return df

# Assuming ekyc_df is the DataFrame containing the 'CUSTOMER_ID' column
#transaction = pd.DataFrame({'CUSTOMER_ID': [generate_random_string(6) for _ in range(10000)]})

# Generate the DataFrame with 444,444 records
trans_df = generate_dataframe(444444, ekyc_df)
```
