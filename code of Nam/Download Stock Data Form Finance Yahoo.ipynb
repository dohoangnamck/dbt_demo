```python
#Import yfinance 
import yfinance as yf 
import matplotlib.pyplot as plt  #for Data visualize
```

```python
# Get the duty for the stock Apple, Google, Facebook, Amazon 
# by specifying the stock ticker, start date, and end date
dataAAPL = yf.download('AAPL', '2022-01-01', '2023-07-01') 
dataGOOG= yf.download('GOOG', '2022-01-01', '2023-07-01') 
dataFB = yf.download('META', '2022-01-01', '2023-07-01') 
dataAMZN = yf.download('AMZN', '2022-01-01', '2023-07-01')
```

```python
#Show data
dataGOOG.head()
```

```python
#Sumpary for Quantitative columns -- Apple inc
dataAAPL.describe()
```

```python
#Sample variance 
dataAAPL.var()          #for all columns
#dataAAPL.Volume.var()    #for Volume columns (khối lượng giao dịch)
```

```python
#Visualize Open and Close price of Apple Inc 
date = dataAAPL.index
plt.plot(date, dataAAPL.Close, label= "Close")  #(x, y) Plot y versus x as lines and/or markers.  
plt.plot(date, dataAAPL.Open, label= "Open")
plt.legend(title="Apple Inc", loc = 'upper left')
fig = plt.gcf()
fig.set_size_inches(15,8)
plt.show()
```

```python
#Visualize Open and Close price of Google Inc 
date = dataGOOG.index
plt.plot(date, dataGOOG.Close, label= "Close")
plt.plot(date, dataGOOG.Open, label= "Open")
plt.legend(title="Google Inc", loc = 'upper left')
fig = plt.gcf()
fig.set_size_inches(15,8)
plt.show()
```

```python
#Visualize Open and Close price of Meta Inc 
date = dataFB.index
plt.plot(date, dataFB.Close, label= "Close")
plt.plot(date, dataFB.Open, label= "Open")
plt.legend(title="Meta Inc", loc = 'upper left')
fig = plt.gcf()
fig.set_size_inches(15,8)
plt.show()
```

```python
#Visualize Open and Close price of Amazon Inc 
date = dataAMZN.index
plt.plot(date, dataAMZN.Close, label= "Close")
plt.plot(date, dataAMZN.Open, label= "Open")
plt.legend(title="Amazon Inc", loc = 'upper left')
fig = plt.gcf()
fig.set_size_inches(15,8)
plt.show()
```

```python
#Visualize Close price of Google, Apple, Fabcebook and Amazon Inc 
date = dataAAPL.index 
plt.plot(date, dataGOOG.Close, label="Google") 
plt.plot(date, dataAAPL.Close, label= "Apple" ) 
plt.plot(date, dataFB.Close, label="Facebook") 
plt.plot(date, dataAMZN.Close, label="Amazon") 
plt.legend(title="Company") 
fig = plt.gcf() 
fig.set_size_inches (20,10) 
plt.show()
```

```python
#Visualize Volume of stock of Google, Apple, Fabcebook and Amazon Inc
date = dataAAPL.index 
plt.plot(date, dataGOOG.Volume, label="Google") 
plt.plot(date, dataAAPL.Volume, label= "Apple" ) 
plt.plot(date, dataFB.Volume, label="Facebook") 
plt.plot(date, dataAMZN.Volume, label="Amazon") 
plt.legend(title="Company") 
fig = plt.gcf() 
fig.set_size_inches (20,10) 
plt.show()
```

```python
#Visualize Volume with boxplot chart 
box_plot_data=[dataGOOG. Volume, dataAAPL. Volume, dataFB.Volume, dataAMZN. Volume] 
fig=plt.gcf() 
fig.set_size_inches(15,10) 
plt.boxplot(box_plot_data, labels =["Google", "Apple", "Facebook ", "Aamazon"]) 
plt.show()
```
