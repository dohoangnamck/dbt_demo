Chúng ta sẽ tính data dữ liệu thêm member_lifespan = expiry_date - join_date

Nên nhớ expiry_date có thể null, trong Python hay SQL thì null tính bất kỳ 1 số gì đó sẽ ra null (sẽ không có lỗi xảy ra 1 số trường hợp)

Trong khi member_lifespan nên được hiểu là tính từ lúc KH đăng ký tới thời điểm tính toán (tức là 2023-01-01), các giá trị null này là 2023-01-01

````python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10, 5))
join_date = pd.to_datetime(df_member['join_date'])
expiry_date_with_null = pd.to_datetime(df_member['expiry_date'])
expiry_date_without_null = pd.to_datetime(df_member['expiry_date'].fillna('2023-01-01'))
result_with_null = ((expiry_date_with_null - join_date) / pd.Timedelta('1 days'))
result_without_null = ((expiry_date_without_null - join_date) / pd.Timedelta('1 days'))
sns.histplot(data=result_with_null, kde=True,ax=ax1,).set(title='Member Lifespan (with null)')
sns.histplot(data=result_without_null, kde=True,ax=ax2).set(title='Member Lifespan (without null)')
plt.show()
````

2 plot rất khác nhau, và cả 2 plot đều có ý nghĩa riêng của nó:
- Hình bên trái thể hiện lifespan của member đã expired, đa phần expired member chỉ tập trung ngắn ngày
- Hình bên phải thể hiện lifespan của toàn bộ member, trong phân phối toàn bộ thì có 2 cụm lớn khách hàng (bao gồm cả member đã expired và còn active)

Đây là ví dụ khác nhau giữa logic business và cách data hoạt động trong thực tế

Nếu bạn không hiểu chính xác sẽ là sai sót trong tính toán

````python
# tạo cột dữ liệu mới member_lifespan
dataframe['member_lifespan'] = result_without_null
````

Trong doanh nghiệp, nhóm member tồn tại ở dưới 3,6,12 tháng sẽ sinh ra vấn đề về customer retention (giữ chân KH)

Ở hình bên trái, đối với member đã expired thì lifespan của họ chỉ tồn tại trong khoảng 30-90 ngày. 

Trường hợp này là 1 ví dụ cho tỷ lệ rời đi (churn-rate) cao. Đây cũng là 1 bài toán khiến các DN nhức đầu, buộc DN phải tung các promotion để thu hút KH dùng sản phẩm. Trong giới hạn nội dung, mình sẽ ko giải quyết bài toán churn-rate mà để các bạn tự tìm hiểu thêm

> **Đọc thêm**: [Customer Retention](https://hocvien.haravan.com/blogs/loyalty-marketing/customer-retention-la-gi)

Hãy xem xét rõ hơn churn-rate với từng phân khúc membership-level

> Churn-rate: ở đây là 1 phép tính đơn giản, khi chỉ tính số lượng member đã expire trong phân khúc membership-level

> Các churn-rate khác bao gồm: churn-rate trong tháng/quý/năm, churn-rate từ tháng x tới tháng y, churn-rate trong campaign...

````python
# Group theo membership_level
for g in df_member.groupby('membership_level'):
    # Đếm tổng số KH đã expired bằng cách membership_status
    num_total_member = len(g[1])
    num_expire = (g[1]['membership_status'] == 'Expired').sum()
    rate_expire = round(num_expire / num_total_member * 100,2)
    print(f'Segment {g[0]} has total {num_total_member} customer and {num_expire} expired customer with general churn-rate {rate_expire}%')
    sns.histplot(data=g[1],x='member_lifespan', kde=True)
    plt.show()
````

## Outer, Inner, Left, Right Join




Cả 2 dataframe đều có chung primary-key là *customer_id*, bạn dùng key này để merge với nhau

- Outer Join: nhiều customer_id key nhất có thể ở cả 2 bảng
- Inner Join: chỉ customer_id key chung
- Left/Right: chỉ customer_id key của 1 trong 2 bảng (tùy trái phải là bảng nào)

````python
# Ví dụ về outer join, mình sẽ tạo thêm 1 vài customer_id có ở bảng này mà ko có ở bảng khác và ngược lại
# lấy 5 row đầu tiên để giữ đc cấu trúc row và thay đổi số customer_id thành 5000 tới 5005
df1 = dataframe[:5].copy()
df1.loc[:,'customer_id']  = list(range(5000,5005))
# gắn vào dataframe cũ
new_dataframe = pd.concat((dataframe, df1), ignore_index=True)
new_dataframe
````

Các bạn thấy được new_dataframe xuất hiện thêm 5 row mới ở sau với customer_id là 5000-5005

Tương tự với member, với 10 row mới

````python
df2 = df_member[:10].copy()
df2.loc[:,'customer_id']  = list(range(6000,6010))
new_df_member = pd.concat((df_member, df2), ignore_index=True)
new_df_member
````

### Outer Join

DataFrame (x) mới có 
- 26 columns (26 = 13+14 - 1(cột chung))
- 1015 row (do có 5 row riêng bảng demographic và 10 row riêng của bảng member)

Các cột ko có thông tin thì sẽ để null (NaT/NaN đều là null trong pandas)

````python
x = pd.merge(new_dataframe, new_df_member, how='outer', left_on='customer_id', right_on='customer_id')
x
````
